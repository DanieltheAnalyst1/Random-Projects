# Prediction

Changing Sales channel to arranged numbers

```python
capstone["Sales Channel"].unique()
```

```python
capstone["Sales Channel"] = capstone["Sales Channel"].map({"Online":0,"Wholesale":1,"In-Store":2,"Distributor":3})
```

```python
capstone["Sales Channel"]
```

```python
import numpy as np 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

x = capstone[['Sales Month', 'Sales Channel', 'Order qty', 'unit price', 'Revenue','unit cost']] 
y = capstone['Profit']

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)

 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
x_train
```

Train the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(x_train,y_train)
```

Test the model

```python
model.predict(x_test)
```
